```python
def count_vowels_and_consonants(string):

    vowel = 'aeiouAEIOU'
    consonants = 'bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ'
    

    vowel_count = 0
    consonant_count = 0
    

    for char in string:
        if char in vowels:
            vowel_count =+ 1
        elif char in consonants:
            consonant_count =+ 1
    
    return vowel_count, consonant_count

input_string = input("Enter a string:  ")


vowels, consonants = count_vowels_and_consonants(input_string)


print(f"Number of vowels:  {vowels}")
print(f"Number of consonants:  {consonants}")
```

```python
def input_list(name):
    
    print(f"Enter the elements of list {name} separated by spaces:")
    elements = list(map(int, input().split()))
    return elements

def main():
    try:
        
        list1 = input_list('List 1')
        list2 = input_list('List 2')

        
        set1 = set(list1)
        set2 = set(list2)

        
        common_elements = set1.intersection(set2)

        
        common_count = len(common_elements)

        print(f"Number of common elements: {common_count}")
    
    except ValueError as ve:
        print(f"Input error: {ve}")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
def input_matrix():

    print("Enter the dimensions of the matrix (rows columns):")
    rows, cols = map(int, input().split())
    
    matrix = []
    print(f"Enter the elements of the matrix row-wise:")
    for _ in range(rows):
        row = list(map(int, input().split()))
        if len(row) != cols:
            raise ValueError(f"Each row must have {cols} columns.")
        matrix.append(row)
    
    return matrix, rows, cols

def transpose_matrix(matrix, rows, cols):

    transpose = [[0] * rows for _ in range(cols)]
    
    for i in range(rows):
        for j in range(cols):
            transpose[j][i] = matrix[i][j]
    
    return transpose

def print_matrix(matrix):
    
    for row in matrix:
        print(' '.join(map(str, row)))

def main():
    try:

        matrix, rows, cols = input_matrix()
        

        transposed = transpose_matrix(matrix, rows, cols)
        
        print("Transpose of the matrix:")
        print_matrix(transposed)
    
    except ValueError as ve:
        print(f"Input error: {ve}")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```
